# Simple Bank

```python
import random
all_accounts=[]
```

```python
all_accounts
```

```python
def open_account(cnic, initial_deposit=0):
    account={'cnic':cnic, 'balance':initial_deposit}
    account_title = input("Enter title of account: ")
    account_number = random.randint(10000,99999)
    account_pin = random.randint(1000,9999)
    account['title'] = account_title
    account['pin'] = account_pin
    account['account_number'] = account_number
    all_accounts.append(account)
    return f"""Thank you for opening account with us!
            {account}"""   
```

```python
print(open_account(54311234, 5000))
```

```python
def check_balance(account_number, pin):
    for account in all_accounts:
        if account['account_number']==account_number:
            if account['pin']==pin:
                return f"Your account balance is Rs:{account['balance']}"
            else:
                return "invalid pin"                
    else:
        print("invalid account number")
```

```python
check_balance(93058,2621)
```

```python
def withdrawl_amt(account_number, pin, amount):
    for account in all_accounts:
        if account['account_number']==account_number:
            if account['pin']==pin:
                if account['balance']>=amount:
                    account['balance']-=amount
                    return f"You withdrawl Rs:{amount} Successfully!"
                else:
                    return "Insufficient Amount"
            else:
                return "Invalid"             
        
    else:
        return "Invalid Account"    
```

```python
withdrawl_amt(93058,2628,2000)
```

```python
check_balance(93058,2628)
```
