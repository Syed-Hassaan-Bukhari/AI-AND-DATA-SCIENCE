Q20: Create a simple calculator:
● Take two numbers
● Take operator (+, -, *, /)
● Perform operation based on user choice

```python
num_a = float(input("Pehla number: "))
num_b = float(input("Doosra number: "))
operator = "/"  
```

```python
print(f"\n--- Calculator ---")
print(f"Numbers: {num_a} {operator} {num_b}")
```

```python
if operator == "+":
    print("Result:", num_a + num_b)
elif operator == "-":
    print("Result:", num_a - num_b)
elif operator == "*":
    print("Result:", num_a * num_b)
elif operator == "/":
    if num_b != 0:
        print("Result:", num_a / num_b)
    else:
        print("Error: Zero se divide nahi ho sakta!")
else:
    print("Galat operator!")
```
