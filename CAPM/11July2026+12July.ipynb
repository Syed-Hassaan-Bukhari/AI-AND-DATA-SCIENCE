```python
import pandas as pd
import numpy as np 
import yfinance as yf
import pandas_datareader
import matplotlib.pyplot as plt
```

```python
apple = yf.download('AAPL', start='2025-07-10', end = '2026-07-10')
google = yf.download('GOOGL', start='2025-07-10', end = '2026-07-10')
```

```python
apple_close = apple['Close']
google_close= google['Close']
```

```python
plt.plot(apple_close)
```

```python
plt.figure(figsize=(16,4))
plt.plot(apple_close)
```

```python
plt.figure(figsize=(22,4))
plt.plot(apple_close)
plt.plot(google_close)
plt.grid()
plt.show()
```

```python
weekly_apple_data = apple_close.resample('W').mean()
weekly_google_data = google_close.resample('W').mean()
```

```python
plt.plot(weekly_apple_data)
plt.plot(weekly_google_data)
plt.grid()
plt.show()
```

```python
import pandas as pd
import yfinance as yf

stocks = ['AAPL', 'GOOGL', 'AMZN', 'MSFT', 'NVDA']

closing_price = pd.DataFrame()

for stock in stocks:
    closing_price[stock] = yf.download(stock,start='2025-07-10',end='2026-07-10')['Close']
closing_price
```

```python
plt.figure(figsize=(22,4))

plt.plot(apple_close, 'b--', label='Apple')   # Blue dashed line
plt.plot(google_close, 'r--', label='Google') # Red dashed line

plt.legend()      # Labels dikhane ke liye
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(22,6))

plt.plot(closing_price)
plt.title("Stock Closing Prices")
plt.xlabel("Date")

plt.legend(stocks)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(22,6))

plt.plot(closing_price['AAPL'], label='Apple')
plt.plot(closing_price['GOOGL'], label='Google')
plt.plot(closing_price['AMZN'], label='Amazon')
plt.plot(closing_price['MSFT'], label='Microsoft')
plt.plot(closing_price['NVDA'], label='NVIDIA')

plt.title("Stock Closing Prices")

plt.legend()
plt.grid(True)

plt.show()
```
