#### Drop the missing data

```python
data.dropna(axis=0) # rows
```

```python
data.dropna(axis=1) # drop columns
```

```python
data.dropna(axis=1, how='all')
```

```python
int(data.shape[0]*0.90)
```

```python
data.dropna(axis=1, thresh=int(data.shape[0]*0.90))
```

```python
data.dropna(axis=0, thresh=int(data.shape[1]*.95))
```

#### Fill the missing data

```python
data.Age.fillna(data.Age.mean())
```

```python
data.Department.fillna(data.Department.mode()[0])
```

```python
data.Department.mode()[0]
```

```python
data.Salary.fillna(data.Salary.mean())
```

```python
data.fillna({'Age':data.Age.mean(), 'Department':data.Department.mode()[0], 'Salary':data.Salary.mean()})
```
