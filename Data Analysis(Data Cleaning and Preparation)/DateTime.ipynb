# Time Series Basics 

```python
dates = [datetime(2011, 1, 2),
         datetime(2011, 1, 5),
         datetime(2011, 1, 7),
         datetime(2011, 1, 8),
         datetime(2011, 1, 10), 
         datetime(2011, 1, 12)]
```

```python
import numpy as np
```

```python
ts = pd.Series(np.random.random(6), index =dates)
```

```python
ts
```

```python
longer_ts = pd.Series(np.random.random(1000), index=pd.date_range('1/1/2000', periods=1000))
longer_ts
```
