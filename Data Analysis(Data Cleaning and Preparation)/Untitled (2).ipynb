# String to date

```python
value = '2011-01-03'
datetime.strptime(value, '%Y-%m-%d')
```

```python
datestrs = ['7/6/2011', '8/6/2011']
```

```python
ts = pd.Series(np.random.randn(6), index=dates)
```

```python
ts
```

```python
longer_ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
longer_ts
```
