```python
class Student():
    def __init__(self,name,roll,course,email, marks):
#         instance variable
#         public
        self.name = name
        self.roll =roll
        self.course=course
        self.email =email
        self.marks = marks
    def show_profile(self):
        print("Student Profile")
        print(f"{self.name}")
        print(f"{self.roll}")
        print(f"{self.course}")
        print(f"{self.email}")
        print(f"{self.marks}")  
    def get_marks(self):
        return self.marks
    def set_marks(self,new_marks):
        if (new_marks>=0) and (new_marks<=100) :
            self.marks = new_marks
        else:
            return 'invalid marks'
```

# Access Modifiers:
    - Public
    - Private
    - Protected

In Python, access modifiers are conventions rather than strict rules. Unlike languages like Java or C++, Python does not use keywords to enforce access restrictions; instead, it uses naming prefixes to signal how attributes and methods should be used.
# Types of Access Modifiers:
  #### Public (Default):Naming: 
         No prefix (e.g., self.name).
         Access: Accessible from anywhere—inside the class, in subclasses, and outside the class.
         Purpose: Used for data and methods intended to be part of the object's public interface.
  #### Protected:Naming: 
        Single underscore prefix (e.g., self._name).
        Access: Accessible within the class and its subclasses. While Python does not physically block external access, the             single underscore is a "gentleman’s agreement" to treat it as non-public.
        Purpose: Alerts other developers that the member is internal and should not be modified directly from outside the class.
  ####  Private:Naming: 
        Double underscore prefix (e.g., self.__name).
        Access: Intended to be accessible only within the class that defines it.Name Mangling: Python enforces this by renaming 
        the member internally to _ClassName__membername to prevent accidental access or name clashes in inheritance.Note: You           can still access these from outside using the mangled name (e.g., obj._ClassName__name), but this is highly            
        discouraged.Comparison TableTypeNaming ConventionInternal AccessSubclass AccessExternal AccessPublicmemberYesYes
